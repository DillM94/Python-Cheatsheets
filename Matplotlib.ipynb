```python
import matplotlib.pyplot as plt
```

# Creating a line plot

```python
x = [0,1,2,3,4,5]
y = [i**2 for i in x]

plt.plot(x,y)
plt.show()
```

```python
# Adding a new line plot:

x = [0,1,2,3,4,5]
y = [i**2 for i in x]
z = [(i**3)*(1/2) for i in x]

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# Getting fancy:

x = [0,1,2,3,4,5]
y = [i**2 for i in x]
z = [(i**3)*(1/2) for i in x]

plt.plot(x,y, color = 'red', linestyle = ':', marker = 'o')
plt.plot(x,z, color = 'green', linestyle = '--', marker = 's')
plt.legend(['X/Y', 'X/Z'])
plt.show()
```

```python
# Shaded Error
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]

plt.plot(months, revenue)
ax = plt.subplot()
ax.set_xticklabels(month_names)
ax.set_xticks(months)

y_lower = [0.9*i for i in revenue]
y_upper = [1.1*i for i in revenue]

plt.fill_between(months, y_lower, y_upper, alpha = 0.2)
plt.show()
```

```python
## Colors:
'''
r: red
b: blue
g: green
etc.... Just type a color it will probably work
'''

## Linestyle:
'''
- solid
-- dashed
-. dased dotted
: dotted
'''

## Marker:
'''
'o' : Circle
'*' : Star
'.' : Point
',' : Pixel
'x' : X
'X' : X (filled)
'+' : Plus
'P' : Plus (filled)
's' : Square
'D' : Diamond
'd' : Diamond (thin)
'p' : Pentagon
'H' : Hexagon
'h' : Hexagon
'v' : Triangle Down
'^' : Triangle Up
'<' : Triangle Left
'>' : Triangle Right
'1' : Tri Down
'2' : Tri Up
'3' : Tri Left
'4' : Tri Right
'|' : Vline
'_' : Hline
'''

## Legend: 
'''
0 - 10: Various locations
'''
```

# Bar Chart

```python
# Using range
heights = [2, 3, 5, 1, 8, 2]
x = range(len(heights))

plt.bar(x,heights)
plt.show()
```

```python
# Using categories
name = ['Dillon', 'Sereena']
years_alive = [26, 24]

plt.bar(name, years_alive)
plt.show()
```

```python
# Side by side bars
'''
n = Number of datasets
t = Dataset number
d = Number of bars
w = Width of each bar

[t*element + w*n for element in range(d)]

This equation generates it each time.
'''
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

n = 1
t = 2
d = 6
w = 0.8

store1_x = [t*element + w*n for element in range(d)]
print(store1_x)
n = 2
t = 2
d = 6
w = 0.8

store2_x = [t*element + w*n for element in range(d)]
print(store2_x)
plt.bar(store1_x, sales1)
plt.bar(store2_x, sales2)

plt.show()
```

```python
# Stacked Bars
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

plt.bar(range(len(drinks)), sales1)
plt.bar(range(len(drinks)), sales2, bottom = sales1)
plt.show()
```

```python
# Error Bars
values = [91, 76, 56, 66, 52, 27]
error = 5

plt.bar(range(len(values)), values, yerr = error, capsize = 5)
plt.show()
```

```python
values = [91, 76, 56, 66, 52, 27]
error = [5, 1, 3, 5, 2, 1]

plt.bar(range(len(values)), values, yerr = error, capsize = 5)
plt.show()
```

# Subplots

```python
x = list(range(0,100))
y = [i**2 for i in x]
z = [i**(1/2) for i in x]

plt.subplot(2,2,1)  ## 2r, 2c, index 1
plt.plot(x,y)

plt.subplot(2,1,2) ## 2r, 1c, index 2
plt.plot(x,z)

plt.subplot(2,2,2) ## 2r, 2c, index 2
plt.plot(y,z)

plt.show()
```

```python
# Creating Space
x = list(range(0,100))
y = [i**2 for i in x]
z = [i**(1/2) for i in x]

plt.subplot(2,2,1)  ## 2r, 2c, index 1
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,1,2) ## 2r, 1c, index 2
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('z')

plt.subplot(2,2,2) ## 2r, 2c, index 2
plt.plot(y,z)
plt.xlabel('y')
plt.ylabel('z')


# Adjusting:
plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
'''
left: left margin
right: right margin
bottom: bottom margin
top: top margin

hspace: Increases vertical space
wspace: Increases horizontal space
'''
plt.show()
```

```python
#2
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]

ax = plt.figure(figsize = (10, 8))
msa_x = create_x(2, 0.8, 1, 5)
msb_x = create_x(2, 0.8, 2, 5)

plt.bar(msa_x, middle_school_a)
plt.bar(msb_x, middle_school_b)

ax = plt.subplot()
ax.set_xticklabels(unit_topics)
ax.set_xticks([(x + y)/2 for (x, y) in zip(msa_x, msb_x)])
plt.axis([0, 10.5, 70, 90])
plt.legend(['Middle School A', 'Middle School B'])
plt.xlabel('Unit')
plt.ylabel('Test Average')
plt.title('Test Averages on Different Units')
plt.show()
```

```python
#3

import numpy as np

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

c_bottom = np.add(As, Bs)

d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)


ax = plt.figure(figsize = (10, 8))
plt.bar(range(len(unit_topics)), As)
plt.bar(range(len(unit_topics)), Bs, bottom = As)
plt.bar(range(len(unit_topics)), Cs, bottom = c_bottom)
plt.bar(range(len(unit_topics)), Ds, bottom = d_bottom)
plt.bar(range(len(unit_topics)), Fs, bottom = f_bottom)

ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)
plt.xlabel('Unit')
plt.ylabel('Number of Students')
plt.title('Grade Distribution')
plt.legend(['A', 'B', 'C', 'D', 'F'])

plt.show()
```
